```python
def buy(transaction_id, date):
    # buy_amount = random.randint(1, 20)
    
    def buy_amount():
        b_range = [1, 3, 6, 8, 11, 14, 20]
        max_b = random.choices(b_range, weights=(0, 23, 20, 25, 17, 13, 2), k=1)[0]
        amount = random.randint(b_range[b_range.index(max_b)-1], max_b)
        
        return amount
        
    adjust = 1 + ((int(date.strftime('%Y%m%d')) - min_date) / (max_date-min_date))
    
    buy_amount = int(buy_amount() * adjust)
                     
    
    cate = [1, 2, 3, 4]
    cost = 0
    items = []
    for _ in range(buy_amount):
        cate_pick = random.choices(cate, weights=(47, 28, 10, 15), k=1)[0]
        item_list = item[item["category_id"] == cate_pick]
        item_pick = item_list.sample(1)
        cost += item_pick["unit_price"].values[0]
        items.append(item_pick["item_id"].values[0])
    
    i = dict(Counter(items))
    transaction = pd.DataFrame({"transaction_id":[],
                    "item_id":[],
                    "item_quantity":[]}, dtype=int)
    
    for k, v in i.items():
        t = pd.DataFrame({"transaction_id":transaction_id,
                         "item_id": [k],
                         "item_quantity": [v],
                         }, dtype=int)

        transaction = pd.concat([transaction, t])
        
    return cost, transaction
```

```python
num = 12332
t_id = [x for x in range(1, num+1)]
transaction_date = sorted([fake.date_between(start_date='-3y', end_date='today') for x in range(num)])
min_date = int(min(transaction_date).strftime('%Y%m%d'))
max_date = int(max(transaction_date).strftime('%Y%m%d'))

customers = [get_customer() for x in range(num)]
```

```python
total = []

transaction_details = pd.DataFrame({"transaction_id":[], "item_id":[], "item_quantity":[]}, dtype=int)

for i in tqdm(range(num)):
    c, t = buy(i+1, transaction_date[i])
    total.append(c)
    transaction_details = pd.concat([transaction_details, t])
```

```python
 def discount(price):
    r = random.random()
    dis = 0
    if r > 0.8:
        dis =  price * random.randint(5, 20) / 100
        
    return dis
```

```python
discount = [ discount(i) for i in total]
```
